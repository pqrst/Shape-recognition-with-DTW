```python
df = pd.read_csv('5z.csv',index_col='timeStamp')
df1 = pd.read_csv('4z.csv',index_col='timeStamp')
```

```python
df['xMA'] = pd.rolling_mean(df['x'],50)
df1['xMA'] = pd.rolling_mean(df1['x'],50)
```

```python
df['yMA'] = pd.rolling_mean(df['y'],50)
df1['yMA'] = pd.rolling_mean(df1['y'],50)
```

```python
df['zMA'] = pd.rolling_mean(df['z'],50)
df1['zMA'] = pd.rolling_mean(df1['z'],50)
```

```python
df.plot()
df1.plot()
plt.show()
```
